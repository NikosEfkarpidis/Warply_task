```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import datetime
import scipy

# read the dataset
data = pd.read_csv(r"C:\Users\nikos\OneDrive\Υπολογιστής\salesdata.csv")

# first look
print(data.columns)
print(data.head())
print(data['sales'].describe())
print(data['customer_id'].value_counts())
print(str(data['ts']))

# change ts from object to datetime , format='%Y-%m-%d%H%M%S.%f%z'
data['dates'] = pd.to_datetime(data['ts'], infer_datetime_format=True)
print(str(data['dates']))
# creation of new var: change the negative values into positive assuming that the inputs are correct as to the absolute value
data['AbsSales'] = data['sales'].abs()
print(data['AbsSales'].describe())

from scipy.stats import skew
# Skewness > 0: Then more weight in the left tail of the distribution.
skew(data['AbsSales'])
from scipy.stats import kurtosis
# For a distribution having kurtosis > 3, It is called leptokurtic
# and it signifies that it tries to produce more outliers rather than the normal distribution.
kurtosis(data['AbsSales'])

# outliers detection and removment
# Creating plot
plt.boxplot(data['AbsSales'])
# show plot
plt.show()

# Detection
# IQR
# Calculate the upper and lower limits
Q1 = data['AbsSales'].quantile(0.25)
Q3 = data['AbsSales'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

# Create arrays of Boolean values indicating the outlier rows
upper_array = np.where(data['AbsSales'] >= upper)[0]
lower_array = np.where(data['AbsSales'] <= lower)[0]

# Removing the outliers
data.drop(index=upper_array, inplace=True)
print(data.shape)

# descriptive
print(data['AbsSales'].describe())
# the maximum seems to be outlier considering the other descriptive values

# histogram
plt.hist(data['AbsSales'])
plt.show()

# Skewness > 0: Then more weight in the left tail of the distribution.
skew(data['AbsSales'])
# For a distribution having kurtosis < 3: It is called playkurtic.
kurtosis(data['AbsSales'])

# Creating box-plot
plt.boxplot(data['AbsSales'])
# show plot
plt.show()
# from boxplot observed that after the cleaning of data there are some new outliers as the data changed

# considering the boxplot removing manually the outliers
out = np.where((data['AbsSales']>488))
data.reindex() # it's not working (i can't understand why) so the next line outputs error
data.drop(index=out[0], inplace=True, axis=1)

# pivot tables with customer and AbsSales
table1 = pd.pivot_table(data,index='customer_id',values='AbsSales',aggfunc=['count','sum','mean'])
print(table1)

# pivot table with customer and mean of  'AbsSales'
table2 = pd.pivot_table(data,index='customer_id',values='AbsSales')
print(table2)
table2.plot(kind='bar')

# pivot table with dates and 'AbsSales'
table3 = pd.pivot_table(data,index='dates',values='AbsSales',aggfunc=['count','sum','mean'])
print(table3)

# pivot table with dates and mean of 'AbsSales'
table4 = pd.pivot_table(data,index='dates',values='AbsSales')
print(table4)
table4.plot(kind='bar')

# creation new variable " DayOfWeek" (1: Monday - 7: Sunday)
data['DayOfWeek'] = data['dates'].dt.isocalendar().day

# pivot table with weekdays and sum of  'AbsSales'
table5 = pd.pivot_table(data,index='DayOfWeek',values='AbsSales',aggfunc='sum')
print(table5)
table5.plot(kind='bar')

# pivot table with weekdays and mean of 'AbsSales'
table6 = pd.pivot_table(data,index='DayOfWeek',values='AbsSales')
print(table6)
table6.plot(kind='bar')

# plots (hist,bar) for date-abssales
def plot_df(df, x, y, title="", xlabel='Date', ylabel='sales', dpi=100):
    plt.figure(figsize=(15,4), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()
plot_df(data,x=data['dates'], y=data['AbsSales'])

def barplot_df(df, x, y, title="", xlabel='dates', ylabel='sales', dpi=100):
    plt.figure(figsize=(15,4), dpi=dpi)
    plt.bar(x, y)
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()
barplot_df(data,x=data['dates'], y=data['AbsSales'])

# plots (hist,bar) for customer_id-abssales
def plot_df1(df, x, y, title="", xlabel='Customer_ID', ylabel='Sales', dpi=100):
    plt.figure(figsize=(15,4), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()
plot_df1(data,x=data['customer_id'], y=data['AbsSales'])

def barplot_df1(df, x, y, title="", xlabel='customer_id', ylabel='sales', dpi=100):
    plt.figure(figsize=(15,4), dpi=dpi)
    plt.bar(x, y)
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()
barplot_df1(data,x=data['customer_id'], y=data['AbsSales'])


# # # take random sample from data for example 10% of data and execute the above code # # #
data1 = data.sample(frac=0.1,replace=False)

# descriptive
print(data1['AbsSales'].describe())

# histogram
plt.hist(data1['AbsSales'])
plt.show()

# Skewness > 0: Then more weight in the left tail of the distribution.
skew(data1['AbsSales'])
# For a distribution having kurtosis < 3: It is called playkurtic.
kurtosis(data1['AbsSales'])

# Creating box-plot
plt.boxplot(data1['AbsSales'])
# show plot
plt.show()

# pivot tables with customer and AbsSales
t1 = pd.pivot_table(data1,index='customer_id',values='AbsSales',aggfunc=['count','sum','mean'])
print(t1)

# pivot table with customer and mean of  'AbsSales'
t2 = pd.pivot_table(data1,index='customer_id',values='AbsSales')
print(t2)
t2.plot(kind='bar')

# pivot table with dates and 'AbsSales'
t3 = pd.pivot_table(data1,index='dates',values='AbsSales',aggfunc=['count','sum','mean'])
print(t3)

# pivot table with dates and mean of 'AbsSales'
t4 = pd.pivot_table(data1,index='dates',values='AbsSales')
print(t4)
t4.plot(kind='bar')

# pivot table with weekdays and sum of  'AbsSales'
t5 = pd.pivot_table(data1,index='DayOfWeek',values='AbsSales',aggfunc='sum')
print(t5)
t5.plot(kind='bar')

# pivot table with weekdays and mean of 'AbsSales'
t6 = pd.pivot_table(data1,index='DayOfWeek',values='AbsSales')
print(t6)
t6.plot(kind='bar')

plot_df(data1,x=data1['dates'], y=data1['AbsSales'])
barplot_df(data1,x=data1['dates'], y=data1['AbsSales'])
plot_df1(data1,x=data1['customer_id'], y=data1['AbsSales'])
barplot_df1(data1,x=data1['customer_id'], y=data1['AbsSales'])
```
